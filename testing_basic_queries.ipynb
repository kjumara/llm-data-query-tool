# Testing Basic Pandas Queries

### Import and Load File

```python
import pandas as pd

file = "sales_data_sample.csv"

for encoding in ['utf-8', 'ISO-8859-1', 'latin1']:
    try:
        df = pd.read_csv(file, encoding=encoding)
        print(f"Data loaded successfully with encoding: {encoding}.")
        break
    except UnicodeDecodeError:
        print(f"Failed to load data with encoding: {encoding}")
        print("Trying next encoding...")
    except FileNotFoundError:
        print(f"File not found at path: {file}")
        df = pd.DataFrame()
        break
    except Exception as e:
        print(f"Error loading data: {str(e)}")
        df = pd.DataFrame()
        break
    print(f"All encodings failed. Returning empty DataFrame.")

df.head()
```

### Convert ORDERDATE to datetime format

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'],errors='coerce')
df.head()
```

### Filter Rows by Product Line


```python
# Available products to change filter to
print("\nAvailable Product Lines:")
product_lines = sorted(df["PRODUCTLINE"].dropna().unique())
for index, item in enumerate(product_lines, start=1):
    print(f"{index}. {item}")
```

```python
product_filter = 'Classic Cars' # Change this variable to filter by other products
filtered_product_df = df[df['PRODUCTLINE'] == product_filter]
filtered_product_df.head()
```

### Filter Rows by Date Range

```python
start_date = '2003-01-01'
end_date = '2003-12-31'

filtered_dates = df[(df['ORDERDATE'] >= start_date) & (df['ORDERDATE'] <= end_date)]
filtered_dates.head()
```

### Total Sales by Product Line

```python
sales_by_product = df.groupby('PRODUCTLINE')['SALES'].sum()
sales_by_product
```

### Sort Products by Total Sales Descending

```python
sales_by_product.sort_values(ascending=False)
```
